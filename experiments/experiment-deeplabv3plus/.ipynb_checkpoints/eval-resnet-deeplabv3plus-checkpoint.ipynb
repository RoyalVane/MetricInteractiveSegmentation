```python
def _fast_hist(label_true, label_pred, n_class):
        mask = (label_true >= 0) & (label_true < n_class)
        hist = np.bincount(
            n_class * label_true[mask].astype(int) + label_pred[mask],
            minlength=n_class ** 2,
        ).reshape(n_class, n_class)
        return hist
```

```python
def compute_miou_acc(model, valloader, n_classes=21):
    n_class = n_classes
    batch_count = 0
    pred_count = 0
    acc = 0.
    cf_matrix = np.zeros(shape=(n_class, n_class), dtype=np.float64)
    
    model = model.eval()
    total_loss = 0.0
    
    with torch.no_grad():
        for i, l in valloader:
            i = i.cuda()
            
            out =  model(i)
            
            out = out.data.cpu().numpy()
            
            out = out.transpose(0,2,3,1)
            out = np.argmax(out,axis = 3)
            
            out = out.reshape(-1)
            lbl = l.numpy().reshape(-1)
            
            assert lbl.shape[0] == out.shape[0]

            indices = np.nonzero(lbl != 255)
            lbl = lbl[indices]
            out = out[indices]

            acc += np.sum(out == lbl)
            pred_count += lbl.shape[0]
            batch_count += 1

            cf_matrix += _fast_hist(lbl, out, n_class)

            del out

    acc /= pred_count
    cf_matrix /= np.sum(cf_matrix)
    hist = cf_matrix
    mean_iou = np.nanmean(np.diag(hist) / (hist.sum(axis=1) + hist.sum(axis=0) - np.diag(hist)))
    
    return mean_iou, acc
```

```python
%%time
miou, acc = compute_miou_acc(model, valloader, n_classes=21)
```

```python
def update_lr_poly(optimizer, init_lr, step, max_step, power):
    lr = init_lr * ((1 - step / float(max_step)) ** (power))
    optimizer.param_groups[0]['lr'] = lr
    optimizer.param_groups[1]['lr'] = 10 * lr

    return optimizer
```

```python
import deeptriplet
import deeptriplet.triplet
```

```python
trainset = deeptriplet.datasets.PascalDatasetRandomTriplet(pascal_root="/scratch/yardima/datasets/voc12/VOCdevkit/VOC2012",
                        split_file="/home/yardima/Python/experiments/pascal_split/train_obj.txt",
                        n_triplets=25000,
                        normalize_imagenet=True,
                        augment=True,
                        pad_zeros=True,
                        downsample_label=1,
                        scale_low=0.8,
                        scale_high=1.2)

trainloader = data.DataLoader(trainset,
                                batch_size=8,
                                num_workers=4,
                                shuffle=True)
```

```python
lr_base = 1e-3
wd = 5e-4
momentum = 0.9
poly_power = 0.9
```

```python
train_params = [{'params': model.get_1x_lr_params(), 'lr': lr_base},
                {'params': model.get_10x_lr_params(), 'lr': lr_base * 10}]
```

```python
train_params = [{'params': list(model.children())[0].get_1x_lr_params(), 'lr': lr_base},
                {'params': list(model.children())[0].get_10x_lr_params(), 'lr': lr_base * 10}]
```

```python
optimizer = torch.optim.SGD(train_params, momentum=momentum, weight_decay=wd)
```

```python
loss_fn = deeptriplet.triplet.RandomTripletPreselected(n_batch=8, n_triplets=25000)
```

```python
nEpochs = 2
max_steps = nEpochs * len(trainloader)
```

```python
step = 0
for e in range(nEpochs):
    
    for i,l in trainloader:
        i = i.cuda()
        l = l.cuda()
        
        optimizer.zero_grad()
        
        out = model.forward(i)
        loss = loss_fn.compute_loss(out, l)
        
        loss.backward()
        
        print(loss.item())
        
        optimizer.step()
        
        optimizer = update_lr_poly(optimizer, lr_base, step, max_steps, poly_power)
        
        del out, loss
        
        step += 1
    
```
