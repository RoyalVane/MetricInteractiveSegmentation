```python
import PIL
import PIL.Image
import numpy as np
```

```python
def normalize_image_imagenet(img, channel_order=False):
    if channel_order:
        img[0,:,:] = (img[0,:,:] - 0.485) / 0.229
        img[1,:,:] = (img[1,:,:] - 0.456) / 0.224
        img[2,:,:] = (img[2,:,:] - 0.406) / 0.225
    else:
        img[:,:,0] = (img[:,:,0] - 0.485) / 0.229
        img[:,:,1] = (img[:,:,1] - 0.456) / 0.224
        img[:,:,2] = (img[:,:,2] - 0.406) / 0.225
        
    return img
```

```python
def load_grabcut(i, dataset, root):
    img = np.array(PIL.Image.open(grabcut_root + dataset[i][0]))
    lbl = np.array(PIL.Image.open(grabcut_root + dataset[i][1]))
    
    img = img.astype(np.float32) / 255
    img = normalize_image_imagenet(img)
    img = np.transpose(img, axes=[2, 0, 1])
    
    lbl[lbl == 255] = 1
    lbl[lbl == 128] = 255
    
    return torch.tensor(img), torch.tensor(lbl)
```

```python
from deeptriplet.datasets import PascalDataset
valset = PascalDataset(pascal_root="/srv/glusterfs/yardima/datasets/voc12/VOCdevkit/VOC2012",
                        split_file="../pascal_split/val_obj.txt",
                        normalize_imagenet=True,
                        augment=False,
                        pad_zeros=True,
                        downsample_label=1)

valloader = data.DataLoader(valset,
                                batch_size=1,
                                num_workers=4,
                                shuffle=False)
```

## Generate and save embeddings

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
for i in range(50):
    image, label = load_grabcut(i,dataset, grabcut_root)

    if image.dim() == 3:
        image = image.unsqueeze(0)

    with torch.no_grad():
        image = image.cuda()
        out = net.forward(image)

        embeddings = out.cpu().data.numpy()
        embeddings = np.transpose(embeddings.squeeze(), axes=[1, 2, 0])

        np.save("/srv/glusterfs/yardima/embed_temp/{}.embed.aug".format(i), embeddings)
        np.save("/srv/glusterfs/yardima/embed_temp/{}.label.aug".format(i), label)
```
